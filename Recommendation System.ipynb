```python
df1=df.rename({'User.ID':'UserId','Book.Title':'BookTitle','Book.Rating':'BookRating'},axis=1)
df1
```

```python
df2=df1.iloc[:,1:]
```

```python
user_df3=df2.pivot_table(index='UserId',columns='BookTitle',values='BookRating').reset_index(drop=True)
user_df3
```

```python
user_df3.index=df2.UserId.unique()
user_df3
```

```python
#Impute those NaNs with 0 values
user_df3.fillna(0, inplace=True)
```

```python
user_df3
```

```python
#Calculating Cosine Similarity between Users
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation
```

```python
user_sim = 1 - pairwise_distances( user_df3.values,metric='cosine')
```

```python
user_sim
```

```python
#Store the results in a dataframe
user_sim_df = pd.DataFrame(user_sim)
user_sim_df
```

```python
#Set the index and column names to user ids 
user_sim_df.index = user_df3.index
user_sim_df.columns = user_df3.index
#user_sim_df.index
user_sim_df.columns
```
